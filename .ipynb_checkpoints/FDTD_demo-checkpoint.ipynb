## The following code shows a demo of using FDTD for  planar microwave circuit modeling

| Parameter    | Description |
|--------------|-------------|
| **d_EM**     | Dimensional data (dt, dx, dy, dz) - represents the discretization parameters in time and three spatial dimensions. |
| **Ni_cells** | Interior points (Nxi, Nyi, Nzi) - defines the number of interior points in each spatial dimension. |
| **Ts**       | Gaussian pulse width - specifies the width of the Gaussian pulse used in the simulation. |
| **steps**    | Number of steps - total number of iterative steps in the simulation process. |
| **ports**    | I/O sampling ports - ports used for input/output sampling. |
| **inc_ports**| Ports for incident fields calculation - specific ports used to calculate incident electromagnetic fields. |
| **eps_sub**  | Substrate dielectric - dielectric constant of the substrate material. |
| **sigma**    | Conductive material distribution - specifies the conductivity distribution across the material. |
| **PEC_layer_x** | Metallization pattern - describes the pattern of metal layers along the x-axis. |
| **save_file**| Choose if save the internal results - option to save or not save the simulation results. |
| **file_name**| File name if save internal results - specifies the file name to save the results if enabled. |
| **tune**     | Use low resolution for calculation - option to use lower resolution for the simulation to save computational resources. |


```python
from EM_solver_FDTD import FDTD_EM
FDTD_solver = FDTD_EM()

# Input dimensions, dx, dy, dz, dt
dx = 0.4064e-3 
dy = 0.4233e-3 
dz = 0.265e-3
CFL = 0.99
light_speed = 2.99792458e8
eps_sub = 2.2
dt = CFL/(light_speed)/((1/dx/dx+1/dy/dy+1/dz/dz)*eps_sub)**0.5
d_EM = [dx, dy, dz, dt]

# interior cells
nxi, nyi, nzi = 80, 100, 16 

# add one cell in each direction for ABCs
nx, ny, nz = nxi + 1, nyi + 1, nzi + 1 

Ni_cells = np.array([nxi, nyi, nzi])

# Gaussian excitation : exp( - ((t-t0)/Ts)**2 ) 
Ts = 15e-12 

# time steps to be run
steps = 1024*4 

# Frequency for computing period
f0 = 20e9

# coordinates of port 1 
nxp1, nyp1 = 32-1, 20-1 

# coordinates of port 2 
nxp2, nyp2 = 48-1, 64-1 

# cell for source excitation 
nyexc = 2-1 

port = np.array([nyexc, nxp1, nyp1, nxp2, nyp2])

# incident port
x_p1, x_p2 = 30-1, 36
y_p1, y_p2 = 1-1, 100
z_p1 = 4-1
inc_port = np.array([x_p1, x_p2, y_p1, y_p2, z_p1])

# metallization layer
PEC_layer_x = np.ones((81,101))
PEC_layer_x[15:65, 47:54] = 0
PEC_layer_x[29:35, 0:47] = 0
PEC_layer_x[45:51, 54:100] = 0

# conductive material distribution
sigma = np.zeros((nx,ny,nz))

# parameters for tuning FDTD resolution
tune = 1

# parameters for saving results
file_name = 'temp.h5'
save_file = False
```

```python
# FDTD parameters
params = {
    'd_EM': d_EM,
    'Ni_cells': Ni_cells,
    'Ts': Ts,
    'steps': steps,
    'f0': f0,
    'port': port,
    'inc_port': inc_port,
    'eps_sub': eps_sub,
    'sigma': sigma,
    'PEC_layer_x': PEC_layer_x,
    'file_name': file_name,
    'save_file': save_file,
    'tune': tune
}
```

```python
# Run FDTD
start_time = time.time()
vinc_t1, vp1_t1, vp2_t1 = FDTD_solver.FDTD(**params)
end_time = time.time()

print(end_time-start_time)
```

```python
freq_t1, S11_t1, S21_t1 = FDTD_solver.S_parameter(vinc_t1, vp1_t1, vp2_t1, steps, dt)

N_FFT = 4*steps
plt.plot(freq_t1*1e-9, 20*np.log10(np.abs(S11_t1[:int(N_FFT/2+1)])),label='S11')
plt.plot(freq_t1*1e-9, 20*np.log10(np.abs(S21_t1[:int(N_FFT/2+1)])),label='S21')

plt.axis([0, 20, -60, 0])
plt.legend()
```

## Run FDTD with a lower resolution

```python
# run FDTD with tune = 3
params['tune'] = 3

start_time = time.time()
vinc_t3, vp1_t3, vp2_t3 = FDTD_solver.FDTD(**params)
end_time = time.time()
print(end_time-start_time)
```

```python
# comparison between tune = 1/3

dx = 0.4064e-3*3
dy = 0.4233e-3 *3
dz = 0.265e-3
CFL = 0.99
eps_sub = 2.2

dt_t3 = CFL/(light_speed)/((1/dx/dx+1/dy/dy+1/dz/dz)*eps_sub)**0.5
freq_t3, S11_t3, S21_t3 = FDTD_solver.S_parameter(vinc_t3, vp1_t3, vp2_t3, steps, dt_t3)

N_FFT = 4*steps
plt.figure(figsize=(12, 4))

plt.subplot(1,2,1)
plt.plot(freq_t1*1e-9, 20*np.log10(np.abs(S11[:int(N_FFT/2+1)])),label='S11 tune = 1')
plt.plot(freq_t3*1e-9, 20*np.log10(np.abs(S11_t3[:int(N_FFT/2+1)])),label='S11 tune = 3')
plt.axis([0, 20, -60, 0])
plt.legend()

plt.subplot(1,2,2)
plt.plot(freq_t1*1e-9, 20*np.log10(np.abs(S21[:int(N_FFT/2+1)])),label='S21 tune = 1')
plt.plot(freq_t3*1e-9, 20*np.log10(np.abs(S21_t3[:int(N_FFT/2+1)])),label='S21 tune = 3')

plt.axis([0, 20, -60, 0])
plt.legend()
```
